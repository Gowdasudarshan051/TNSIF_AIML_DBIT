```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Example plot
x = np.linspace(0, 10, 100)
y = np.cos(x)
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show() # While not strictly necessary with %matplotlib inline, it's good practice.
```
